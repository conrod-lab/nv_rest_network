This is a random matrix example to just fix the logic

```python
#create a random matrix which is 5X5 with 0s on the diagonal
random_matrix = numpy.random.rand(5,5)
numpy.fill_diagonal(random_matrix, 0)

random_matrix.shape
```

```python
#make random_matrix a dataframe

random_matrix_df = pd.DataFrame(random_matrix)
#add a new column to the random_matrix_df

random_matrix_df['network'] = [1,1,2,2,3]

```

```python


# Create a dictionary to store indices for each network
network_indices = {}

# Iterate over the unique network values
for network_value in random_matrix_df['network'].unique():
    # Get the indices where 'network' column has the current network value
    indices = random_matrix_df.index[random_matrix_df['network'] == network_value].tolist()
    
    # Extract the row and column indices
    row_indices = [index // len(random_matrix) for index in indices]
    col_indices = [index % len(random_matrix) for index in indices]
    
    # Store the row and column indices in the dictionary
    network_indices[network_value] = list(zip(row_indices, col_indices))

# Print the dictionary
print(network_indices)



# Calculate within-network mean
within_network_means = {}
for network_value, indices in network_indices.items():
    # Extract values corresponding to the indices
    values = [random_matrix[row][col] for row, col in indices]
    # Calculate mean
    within_network_means[network_value] = np.mean(values)

# Calculate between-network mean
between_network_means = {}
for network_value1, indices1 in network_indices.items():
    for network_value2, indices2 in network_indices.items():
        if network_value1 != network_value2:
            # Extract values corresponding to the indices of each network
            values1 = [random_matrix[row][col] for row, col in indices1]
            values2 = [random_matrix[row][col] for row, col in indices2]
            # Calculate mean between networks
            mean_between_networks = np.mean(values1 + values2)
            # Store the mean between networks
            between_network_means[(network_value1, network_value2)] = mean_between_networks

print("Within-network means:")
print(within_network_means)
print("Between-network means:")
print(between_network_means)
```
